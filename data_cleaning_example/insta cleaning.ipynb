```python
import pandas as pd

data = pd.read_excel("data/wonderpyo_insta.xlsx")

data
```

```python
from unicodedata import normalize


for i in range(len(data)):
    data['내용'][i] = normalize('NFC',data['내용'][i])
    data['댓글'][i] = normalize('NFC',data['댓글'][i])
```

```python
import re

hash_tag = re.compile("#[ㄱ-ㅎ가-힣a-zA-Z0-9_]+")
data['해시테그'] = ''

for i in range(len(data)):
    ht = list()
    c = ",".join(hash_tag.findall(data['내용'][i]))
    cc = ",".join(hash_tag.findall(data['댓글'][i]))
    ht.append(c)
    ht.append(cc)

    data['해시테그'][i] = ",".join(ht)
```

```python
data['new댓글'] = ''

for j in range(len(data)):
    dd = data['댓글'][j].split("', '")

    comments = list()
    
    try:
        for i in range(len(dd)):
            comments.append(dd[i].split("\\n")[1])
    except:
        comments.append('댓글없음')
        
    data['new댓글'][j] = ",".join(comments)
```

```python
data = data[['인플루언서','날짜','게시물타입','좋아요수/조회수','내용','해시테그','댓글수','new댓글','url']]

data.rename({'new댓글':'댓글'}, axis = 'columns', inplace = True)
```
